```python
num=[]
for col in ent.columns:
    if(ent[col].dtypes!='object'):
        num.append(col)
print(f"There aret total {len(num)} numerical columns in datasest")
print(num)
```

```python
normal=[]
for col in num:
    if((ent[col].skew()>-0.5) and (ent[col].skew()<0.5)):
        normal.append(col)
print(f"There are total {len(normal)} numerical columns that have normal distribtion")
print(normal)
```

```python
sns.set_style(rc={'figure.facecolor':'#DDF790'})
fig,axes=plt.subplots(2,2,figsize=(10,10))
axes=axes.ravel()
for col,ax in zip(normal,axes):
    sns.kdeplot(ent[col],ax=ax)
plt.show()
```

```python
right_skew=[]
for col in num:
    if(ent[col].skew()>0.5):
        right_skew.append(col)
print(f"There are total {len(right_skew)} right-skewed columns in dataset")
print(right_skew)
```

```python
left_skew=[]
for col in num:
    if(ent[col].skew()<-0.5):
        left_skew.append(col)
print(f"There are total {len(left_skew)} left-skewed columns in dataset")
print(left_skew)
```

```python
sns.set_style(rc={'figure.facecolor':'#F7CD90'})
num_plots = len(left_skew)
fig, ax = plt.subplots(1, 1, figsize=(13, 2))
col = left_skew[0]
sns.kdeplot(ent[col], ax=ax)
plt.tight_layout() 
plt.show()
```

```python
mean=[]
std=[]
for col in normal:
    mean.append(ent[col].mean())
    std.append(ent[col].std())
```

```python
upper_limit=[]
lower_limit=[]

for i in range(len(normal)):
    upper_limit.append(mean[i]+3*std[i])
    lower_limit.append(mean[i]-3*std[i])
```

```python
data_index=[]   
for i in range(len(normal)):
    print("")
    print("")
    print(f"Outlier for{normal[i]}")
    print(f"Upper_limit for {normal[i]} is {upper_limit[i]}")
    print(f"Lower limit for {normal[i]} is {lower_limit[1]}")
    d=ent[normal[i]][(ent[normal[i]]>upper_limit[i]) | (ent[normal[i]]<lower_limit[i])]
    if len(d)>0:
        data_index.append(d.index)
    print(d)
    print(len(d))
print(f"There are total {len(data_index)} outliers in normal distributed columns")
```

```python
per_25=[]
per_75=[]
for col in right_skew:
    per_25.append(ent[col].quantile(0.25))
    per_75.append(ent[col].quantile(0.75))

iqr=[]
for i in range(len(right_skew)):
    iqr.append(per_75[i]-per_25[i])

upper_r=[]
lower_r=[]
for i in range(len(right_skew)):
    upper_r.append(per_75[i]+1.5*iqr[i])
    lower_r.append(per_25[i]-1.5*iqr[i])
```

```python
r_data_index = []

for i in range(len(left_skew)):
    print("")
    print("")
    print(f"Outlier for {left_skew[i]}")
    if i < len(upper_r):
        print(f"Upper_limit for {left_skew[i]} is {upper_r[i]}")
    else:
        print("No upper limit available")
    
    if i < len(lower_r):
        print(f"Lower limit for {left_skew[i]} is {lower_r[i]}")
    else:
        print("No lower limit available")
    if i < len(upper_r) and i < len(lower_r):
        d_r = ent[left_skew[i]][(ent[left_skew[i]] > upper_r[i]) | (ent[left_skew[i]] < lower_r[i])]
        if len(d_r) > 0:
            r_data_index.append(d_r.index)
        print(d_r.index)
        print(len(d_r))
    else:
        print("No outlier data available")

```

```python
import numpy as np

for i in range(len(left_skew)):
    if i < len(upper_r) and i < len(lower_r):
        ent[left_skew[i]] = np.where(ent[left_skew[i]] > upper_r[i], upper_r[i],
                                     np.where(ent[left_skew[i]] < lower_r[i], lower_r[i], ent[left_skew[i]]))
    else:
        print("No upper or lower limit available for", left_skew[i])
```
